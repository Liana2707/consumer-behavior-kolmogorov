```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = [
    pd.read_csv('out_1.csv', index_col=0),
    pd.read_csv('out_2.csv', index_col=0),
    pd.read_csv('out_3.csv', index_col=0)]
data
```

```python
fig, axs = plt.subplots(3, 1, figsize=(4, 14))
n_bins = 20 
categories = ['socialisation', 'survival', 'self_realisation']
palettes = ['RdPu_r', 'YlGnBu_r', 'Reds_r']
for i in range(3):
    sns.histplot(data=data[i]['hp_7'],  ax=axs[i], bins=n_bins)
    axs[i].set_title(categories[i])
```

```python
features = ['hp_7']
fig, axes = plt.subplots(3, 1, figsize=(4, 12))
axes = axes.flatten()

for i in range(3):
    ax = axes[i]
    sns.scatterplot(x=data[i]['lz_complexity'], y=data[i]['hp_7'], ax=ax, hue=data[i]['hp_7'], palette=palettes[i]).legend().remove()
    ax.set_xlabel('lz Complexity')
    ax.set_ylabel('hp_7')
    ax.set_title(f' hp_7     {categories[i]}')

    correlation = data[i]['lz_complexity'].corr(data[i]['hp_7'])
    ax.text(0.1, 0.9, f' : {correlation:.2f}', transform=ax.transAxes)
    ax.grid(True)

plt.tight_layout()
plt.show()
```

```python
features = ['hp_7']
fig, axes = plt.subplots(3, 1, figsize=(4, 12))
axes = axes.flatten()

for i in range(3):
    ax = axes[i]
    sns.scatterplot(x=data[i]['huffman_complexity'], y=data[i]['hp_7'], ax=ax, hue=data[i]['hp_7'], palette=palettes[i]).legend().remove()
    ax.set_xlabel('huffman Complexity')
    ax.set_ylabel(f'hp_7')
    ax.set_title(f' hp_7     {categories[i]}')

    correlation = data[i]['huffman_complexity'].corr(data[i]['hp_7'])
    ax.text(0.1, 0.9, f' : {correlation:.2f}', transform=ax.transAxes)
    ax.grid(True)

plt.tight_layout()
plt.show()
```
